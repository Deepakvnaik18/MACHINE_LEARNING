```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("logistic_regression_dataset.csv")

df.head()
```

```python
df = df.drop(columns=("Category"),axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

x_tfidf = tfidf.fit_transform(df["Review"])

x_tfidf
```

```python
feature_column_names = tfidf.get_feature_names_out()
feature_column_names
```

```python
feature_values = x_tfidf.toarray()
feature_values
```

```python
vector_df = pd.DataFrame(feature_values,columns=(feature_column_names))

vector_df
```
