```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,recall_score
```

```python
df = sns.load_dataset("titanic")

df.head()
```

```python
df.info()
```

```python
x = df.drop(columns="survived")

y = df["survived"]
```

```python
y.value_counts()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
cat_col = x.select_dtypes(include="object").columns

num_col = x.select_dtypes(include=["int64","float64"]).columns
```

```python
num = Pipeline(
    steps=[
        ("num_pre",SimpleImputer(strategy="mean")),
       ( "stand",StandardScaler())
				]
)

cat = Pipeline(
    steps=[
        ("num_pre",SimpleImputer(strategy="most_frequent")),
       ( "stand",OneHotEncoder())
				]
)

preprocessing  = ColumnTransformer(
    transformers=[
        ('pre_num',num,num_col),
        ("pre_col",cat,cat_col)
				]
)
```

# BernoulliNB


```python
pipeline = Pipeline(
    steps=[
        ("preprocessor",preprocessing),
        ("Barnoulli",BernoulliNB())
				]
)
```

```python
pipeline
```

```python
pipeline.fit(xtrain,ytrain)
```

```python
pred = pipeline.predict(xtest)
```

```python
confusion_matrix(ytest,pred)
```

```python
precision_score(ytest,pred,pos_label=0)

precision_score(ytest,pred,pos_label=1)
```

```python
recall_score(ytest,pred,pos_label=0)

recall_score(ytest,pred,pos_label=1)
```

```python
f1_score(ytest,pred,pos_label=1)

f1_score(ytest,pred,pos_label=0)
```

# GaussianNB

```python
Pipeline_1 = Pipeline(
    steps=[
        ("preprocessor",preprocessing),
        ("Barnoulli",GaussianNB())
				]
)
```

```python
Pipeline_1.fit(xtrain,ytrain)
```

```python
Pipeline_1.score(xtest,ytest)
```
